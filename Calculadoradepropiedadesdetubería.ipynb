```python
import gradio as gr
import math

# Funciones para calcular las propiedades de la tubería

def calcular_area_tuberia(Diametro):
    # Fórmula para calcular el área de un círculo: A = π * (d/2)^2
    area = math.pi * ((Diametro * 0.0254) ** 2) / 4
    return area

def calcular_velocidad(Caudal, Diametro):
    area = calcular_area_tuberia(Diametro)
    velocidad = 0.001 * Caudal / area
    return velocidad

def calcular_reynolds(Densidad, Viscosidad, Velocidad, Diametro):
    reynolds = 25.4 * ((Densidad * Velocidad * Diametro) / Viscosidad)
    return reynolds

def calcular_factor_friccion(Diametro, Reynolds):
    ed = 0.0018 / Diametro
    cf = 7
    for i in range(6):
        cf = -2 * math.log10(ed / 3.71 + 2.51 / Reynolds * cf)
    ff = 1 / (cf * cf)
    return ff

def calcular_longitud_total(Longitud_recta, Diametro):
    longitud_recta_metros = Longitud_recta * 1000  # Convertir kilómetros a metros
    longitud_accesorios = longitud_recta_metros * 0.05
    longitud_total = longitud_recta_metros + longitud_accesorios * (Diametro / 39.3701)
    return longitud_total

def calcular_perdidas_carga_friccion(Factor_friccion, Longitud_total, Diametro, Velocidad):
    perdidas_carga_friccion = (Factor_friccion * (Longitud_total / Diametro) * (Velocidad ** 2)) / 0.498348
    return perdidas_carga_friccion

def calcular_presion(Metros_sobre_nivel_mar):
    presion = 100 * 1013.25 * (1 - 0.0000225577 * Metros_sobre_nivel_mar) ** 5.2559
    return presion

def calcular_caida_presion(Metros_sobre_nivel_mar_1, Metros_sobre_nivel_mar_2, Perdidas_carga_friccion):
    presion_1 = calcular_presion(Metros_sobre_nivel_mar_1)
    presion_2 = calcular_presion(Metros_sobre_nivel_mar_2)
    caida_presion = (presion_2 - presion_1) / (1000 * 9.81) + Perdidas_carga_friccion
    return caida_presion

def calcular_deltap(Densidad, Metros_sobre_nivel_mar_1, Metros_sobre_nivel_mar_2, Perdidas_carga_friccion):
    caida_presion = calcular_caida_presion(Metros_sobre_nivel_mar_1, Metros_sobre_nivel_mar_2, Perdidas_carga_friccion)
    deltap = 9.81 * Densidad * caida_presion
    return deltap

def calcular_BHP(Caudal, Deltap):
    BHP = 0.01 * Caudal * Deltap
    return BHP

def calcular_HP(BHP, Eficiencia):
    HP = BHP / (746 * Eficiencia)
    return HP

def calcular_HP_kW(BHP, Eficiencia):
    HP_kW = BHP / (1000 * Eficiencia)
    return HP_kW

def calcular_consumo_energetico(HP_kW):
    consumo_energetico = HP_kW * 8760  # Horas por año
    return consumo_energetico

def calcular_costo_por_anno(consumo_energetico, Precio_kwh):
    costo_por_anno = consumo_energetico * Precio_kwh
    return costo_por_anno

def calcular_volumen(Diferencia_area, Longitud_recta):
    volumen = Diferencia_area * Longitud_recta * 1000  # Convertir a metros cúbicos
    return volumen

def calcular_masa_tuberia(Volumen_tuberia, Densidad_acero):
    masa_tuberia = Volumen_tuberia * Densidad_acero
    return masa_tuberia

def calcular_precio_tuberia(Masa_tuberia, Precio_por_kg):
    precio_tuberia = Masa_tuberia * Precio_por_kg
    return precio_tuberia

def din(dn):
    if dn == 0.125:
        di = 0.269
    elif dn == 0.25:
        di = 0.364
    elif dn == 0.375:
        di = 0.493
    elif dn == 0.5:
        di = 0.622
    elif dn == 0.75:
        di = 0.824
    elif dn == 1:
        di = 1.049
    elif dn == 1.25:
        di = 1.38
    elif dn == 1.5:
        di = 1.61
    elif dn == 2:
        di = 2.067
    elif dn == 2.5:
        di = 2.469
    elif dn == 3:
        di = 3.068
    elif dn == 3.5:
        di = 3.548
    elif dn == 4:
        di = 4.026
    elif dn == 4.5:
        di = 4.506
    elif dn == 5:
        di = 5.047
    elif dn == 6:
        di = 6.065
    elif dn == 8:
        di = 7.981
    elif dn == 10:
        di = 10.02
    elif dn == 12:
        di = 11.938
    elif dn == 14:
        di = 13.124
    elif dn == 16:
        di = 15
    elif dn == 18:
        di = 16.876
    elif dn == 20:
        di = 18.812
    elif dn == 24:
        di = 22.626
    elif dn == 30:
        di = 29.25
    elif dn == 36:
        di = 35.25
    elif dn == 48:
        di = 47.25
    else:
        di = "Hola es un error fuera de rango"
    return di

def calcular_interfaz(Caudal, Densidad, Viscosidad, Dn, Longitud_recta, Metros_sobre_nivel_mar_1, Metros_sobre_nivel_mar_2, Eficiencia, Precio_kwh, Densidad_acero, Precio_por_kg):
    di = din(Dn)
    Diametro = di
    Area = calcular_area_tuberia(Diametro)
    Velocidad = calcular_velocidad(Caudal, Diametro)
    Reynolds = calcular_reynolds(Densidad, Viscosidad, Velocidad, Diametro)
    Factor_friccion = calcular_factor_friccion(Diametro, Reynolds)
    Longitud_total = calcular_longitud_total(Longitud_recta, Diametro)
    Perdidas_carga_friccion = calcular_perdidas_carga_friccion(Factor_friccion, Longitud_total, Diametro, Velocidad)
    Caida_presion = calcular_caida_presion(Metros_sobre_nivel_mar_1, Metros_sobre_nivel_mar_2, Perdidas_carga_friccion)
    Deltap = calcular_deltap(Densidad, Metros_sobre_nivel_mar_1, Metros_sobre_nivel_mar_2, Perdidas_carga_friccion)
    BHP = calcular_BHP(Caudal, Deltap)
    HP = calcular_HP(BHP, Eficiencia)
    HP_kW = calcular_HP_kW(BHP, Eficiencia)
    Consumo_energetico = calcular_consumo_energetico(HP_kW)
    Costo_por_anno = calcular_costo_por_anno(Consumo_energetico, Precio_kwh)
    Volumen_tuberia = calcular_volumen(Area, Longitud_recta)
    Masa_tuberia = calcular_masa_tuberia(Volumen_tuberia, Densidad_acero)
    Precio_tuberia = calcular_precio_tuberia(Masa_tuberia, Precio_por_kg)
    return f"Diámetro interno: {di:.4f} metros\nÁrea de la tubería: {Area:.4f} metros cuadrados.\nVelocidad en la tubería: {Velocidad:.2f} m/s.\nNúmero de Reynolds: {Reynolds:.2f}.\nFactor de fricción: {Factor_friccion:.4f}\nLongitud total: {Longitud_total:.2f} metros\nPérdidas de carga por fricción: {Perdidas_carga_friccion:.4f}\nCaida de presión: {Caida_presion:.4f} metros\nDeltap: {Deltap:.4f} Pascales\nBHP (W): {BHP:.4f}\nHP (kW): {HP_kW:.4f}\nHP: {HP:.4f}\nConsumo energético: {Consumo_energetico:.4f} kWh\nCosto por año: {Costo_por_anno:.4f} MXN\nVolumen de la tubería: {Volumen_tuberia:.4f} metros cúbicos\nMasa de la tubería: {Masa_tuberia:.4f} kilogramos\nPrecio de la tubería: {Precio_tuberia:.4f} MXN"

# Interfaz de usuario
iface = gr.Interface(fn=calcular_interfaz, inputs=["number", "number", "number", "number", "number", "number", "number", "number", "number", "number", "number"], outputs="text",
                     title="Calculadora de Propiedades de Tubería",
                     description="Fernandez Nuñes Karen Ismerai N° de control: 20440676\nOntiveros Atondo Annivianey N° de control: 20440356\nRomero Figueroa José Roberto N° de control: 20440368\nValdez Aguirre Alexia Berenice  N° de control: 20441547\n\nIngresa el Caudal (litros/segundo), la Densidad (kg/m³), la Viscosidad (cP), el Diámetro nominal, la Longitud recta (kilómetros), los Metros sobre el nivel del mar (1), los Metros sobre el nivel del mar (2), la Eficiencia (%) y el Precio del kWh para calcular el área, la velocidad, el número de Reynolds, el factor de fricción, la longitud total, las pérdidas de carga por fricción, la caída de presión, deltap, BHP (W), el consumo energético, el costo por año, el volumen de la tubería, la masa de la tubería y el precio de la tubería. Además, proporciona la Densidad del acero y el Precio por kilogramo de la tubería.")

iface.launch()
```
